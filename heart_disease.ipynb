```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
x=df.drop(columns=["AHD"], axis=1)
```

```python
x
```

```python
y=df["AHD"]
```

```python
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler=StandardScaler()
```

```python
xtrain_scaled=scaler.fit_transform(xtrain)
xtest_scaled=scaler.fit_transform(xtest)
```

```python
model={
    "Logistic Regression": LogisticRegression(),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(),
    "Support Vector Machine": SVC(kernel="rbf"),
    "K-Nearest Neighbors": KNeighborsClassifier()
}
```

```python
result=[]
```

```python
for name,m in model.items():
    m.fit(xtrain_scaled, ytrain)
    ypred=m.predict(xtest_scaled)
    acc=accuracy_score(ytest, ypred)
    f1=f1_score(ytest, ypred)
    result.append({"Model":name, "Accuracy":round(acc,2), "F1 Score":round(f1,2)})
```

```python
result
```

```python
import joblib
joblib.dump(model['Support Vector Machine'],"svm_heart.pkl")
joblib.dump(scaler,"scaler.pkl")
joblib.dump(x.columns.tolist(),"columns.pkl")
```
